```python
import langchain
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import Chroma
```

```python
documnet_loader = TextLoader('/content/Quaideazam.txt')
loader = documnet_loader.load()
loader
```

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=25)
documents = text_splitter.split_documents(loader)

documents
```

```python
Embeddings = HuggingFaceEmbeddings()
hey = "Hello this is zack amrtin from healthcare benfits"
vectors = Embeddings.embed_query(hey)
vectors
```

```python
from langchain_community.vectorstores import FAISS
```

```python
faiss_db = FAISS.from_documents(documents, Embeddings)
faiss_db.save_local("faiss_index")
```

```python
question = "Early political invovement"
```

```python
retriver = faiss_db.as_retriever(search_type="similarity", search_kwargs={"k": 2})
sear = retriver.get_relevant_documents(question)
for i in sear:
  print(i.page_content)
```
