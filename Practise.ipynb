```python
from langchain.prompts import ChatPromptTemplate
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter
from langchain.retrievers import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import LLMChainExtractor
```

```python
textloader = TextLoader("/content/Quaideazam.txt")
data = textloader.load()
data
```

```python
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
texts = text_splitter.split_documents(data)
```

```python
texts
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
vectorStores = FAISS.from_documents(texts,embeddings)
```

```python
retriver = vectorStores.as_retriever(search_type="similarity", search_kwargs={"k":1})
```

```python
compressor  = LLMChainExtractor.from_llm(llm)
```

```python
CompressorChain = ContextualCompressionRetriever(base_compressor=compressor,base_retriever=retriver)
```

```python
res=CompressorChain.get_relevant_documents("write something about quaid e azam?")
```
