```python
from langchain_community.document_loaders import TextLoader
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.vectorstores import FAISS
from langchain.text_splitter import CharacterTextSplitter
from langchain.retrievers import ContextualCompressionRetriever
from langchain.retrievers.document_compressors import LLMChainExtractor
```

```python
text_loader = TextLoader('/content/Quaideazam.txt')
data = text_loader.load()
```

```python
def characterSplitter(text):
  text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=50)
  texts = text_splitter.split_documents(text)
  return texts

text = characterSplitter(data)
```

```python
len(text)
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
vector = embeddings.embed_query("Hello waris")
print(vector)
```

```python
!pip install faiss-cpu
```

```python
VectorStore = FAISS.from_documents(text,embeddings)
```

```python
retriver = VectorStore.as_retriever(search_type="similarity", search_kwargs={"k":1})
```

```python
query = "Jinnah’s leadership of the Muslim League during the 1930s and 1940s"
```

```python
#From langchain Extractor
compressor = LLMChainExtractor.from_llm(llm)

#Contextual Compressor to get state of the art result
Compressor_retriver = ContextualCompressionRetriever(base_compressor=compressor, base_retriever=retriver)
```

```python
result = Compressor_retriver.get_relevant_documents(query)
print(result[0].page_content)
```
